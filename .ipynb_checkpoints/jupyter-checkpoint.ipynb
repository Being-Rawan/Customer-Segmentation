```python
import pandas as pd

# Load the dataset from the Excel file
df = pd.read_excel('C:\\Users\\Rawan Hesham\\OneDrive\\Desktop\\Customer segmentation\\Online Retail.xlsx')

# Show the first few rows of the dataset
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.duplicated().sum()  # This will show how many duplicate rows exist
```

```python
# View Duplicate Rows
df[df.duplicated()]
```

```python
# Overview of the dataset
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Count rows where Quantity is less than or equal to 0
invalid_quantity_rows = df[df['Quantity'] <= 0]
print(f"Number of rows with Quantity <= 0: {len(invalid_quantity_rows)}")

# Display a few rows to inspect
invalid_quantity_rows.head()
```

```python
# Remove transactions where Quantity is less than or equal to 0
df = df[df['Quantity'] > 0]
```

```python
# Count rows where Quantity is less than or equal to 0
invalid_quantity_rows = df[df['Quantity'] <= 0]
print(f"Number of rows with Quantity <= 0: {len(invalid_quantity_rows)}")

# Display a few rows to inspect
invalid_quantity_rows.head()
```

```python
# Count rows where UnitPrice is less than or equal to 0
invalid_unitprice_rows = df[df['UnitPrice'] <= 0]
print(f"Number of rows with UnitPrice <= 0: {len(invalid_unitprice_rows)}")

# Display a few rows to inspect
invalid_unitprice_rows.head()
```

```python
# Remove transactions where UnitPrice is less than or equal to 0
df = df[df['UnitPrice'] > 0]
```

```python
# Count rows where UnitPrice is less than or equal to 0
invalid_unitprice_rows = df[df['UnitPrice'] <= 0]
print(f"Number of rows with UnitPrice <= 0: {len(invalid_unitprice_rows)}")

# Display a few rows to inspect
invalid_unitprice_rows.head()
```

```python
# Check for duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")

# Display duplicates (if any)
duplicates = df[df.duplicated()]
duplicates.head()
```

```python
# Check for duplicate rows (based on invoiceNo and customerID)
df.duplicated(subset=['InvoiceNo', 'CustomerID']).sum()
```

```python
# Remove duplicate InvoiceNo for the same CustomerID
df = df.drop_duplicates(subset=['InvoiceNo', 'CustomerID'])
```

```python
# Check for duplicate rows (based on invoiceNo and customerID)
df.duplicated(subset=['InvoiceNo', 'CustomerID']).sum()
```

```python
# Show the first few rows of the dataset
df.head()
```

```python
# Overview of the dataset
df.info()
```

```python
# Group by CustomerID and count the number of unique InvoiceNo for each customer
purchase_frequency = df.groupby('CustomerID')['InvoiceNo'].nunique()

# Convert the series to a DataFrame for easier viewing
purchase_frequency = purchase_frequency.reset_index()
purchase_frequency.columns = ['CustomerID', 'PurchaseCount']
```

```python
# Print the modified dataset (DataFrame)
print(purchase_frequency)
```

```python
# View the first few rows of the modified dataset
purchase_frequency.head()
```

```python
# Categorize customers based on purchase count
purchase_frequency['FrequencyCategory'] = purchase_frequency['PurchaseCount'].apply(lambda x: 'Low Frequency' if x <= 5 else 'High Frequency')
```

```python
# View the first few rows of the modified dataset
purchase_frequency.head()
```

```python
# Inspect the final dataset
df.head()
```

```python
# Save the preprocessed dataset (purchase_frequency) to a new Excel file
purchase_frequency.to_excel('processed_purchase_frequency.xlsx', index=False)
```
